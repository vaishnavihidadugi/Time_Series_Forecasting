```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Check current working directory
print("Current directory:", os.getcwd())

# List all files in current directory
print("Files in this directory:", os.listdir())

# Path to your CSV file (adjust if needed)
path = "cleaned_aapl_data.csv"

# Check if file exists before loading
if not os.path.isfile(path):
	print(f"File not found at {path}. Please check the path and try again.")
	# Optionally, list files in the parent directory to help debug
	print("Files in '../data':", os.listdir(os.path.dirname(path)))
else:
	# Load dataset
	df = pd.read_csv(path)
	# Preview data
	display(df.head())
```

```python
# Check if df is defined
if 'df' in locals():
	# Ensure Date is datetime
	df['Date'] = pd.to_datetime(df['Date'])

	# Plot Close price over time
	plt.figure(figsize=(14, 6))
	plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
	plt.title('Close Price Over Time')
	plt.xlabel('Date')
	plt.ylabel('Close Price (USD)')
	plt.legend()
	plt.show()
else:
	print("DataFrame 'df' is not defined. Please ensure the data is loaded successfully in the previous cell.")
```
